```python
import cv2
from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
# import pandas as pd
import numpy as np
import math
```

## Menampilkan Histogram

```python
img = cv2.imread('hands (1).jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
hist = cv2.normalize(hist, hist).flatten()

img2 = cv2.imread('hands (2).jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
hist2 = cv2.calcHist([img2], [0], None, [256], [0, 256])
hist2 = cv2.normalize(hist2, hist2).flatten()

img3 = cv2.imread('hands (3).jpg')
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
hist3 = cv2.calcHist([img3], [0], None, [256], [0, 256])
hist3 = cv2.normalize(hist3, hist3).flatten()
```

```python
fig = plt.figure(figsize=(20, 20))
fig.add_subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
fig.add_subplot(1, 3, 2)
plt.imshow(img2, cmap='gray')
fig.add_subplot(1, 3, 3)
plt.imshow(img3, cmap='gray')
plt.show()

fig2 = plt.figure(figsize=(20, 5))
fig2.add_subplot(1, 3, 1)
plt.hist(img.ravel(), 256, [0,256])
fig2.add_subplot(1, 3, 2)
plt.hist(img2.ravel(), 256, [0,256])
fig2.add_subplot(1, 3, 3)
plt.hist(img3.ravel(), 256, [0,256])
plt.show()
```

## Histogram Biner

```python
img = cv2.imread('biner.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
hist = cv2.normalize(hist, hist).flatten()


plt.imshow(img, cmap='gray')
plt.show()
plt.hist(img.ravel(), 256, [0,256])
plt.show()
```

## Histogram Citra RGB

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('hands (1).jpg')
plt.imshow(img)
plt.show()
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

## Membandingkan Histogram
- Correlation: A perfect match is 1 and a maximal mismatch is –1
- Chi-Square: A perfect match is 0 and a total mismatch is unbounded (depending on the size of the histogram).
- Intersection: A perfect match is 1 and a total mismatch is 0
- Bhattacharyya: A perfect match is 0 and a total mismatch is a 1. 

```python
print('Citra 1 dengan Citra 2')
print('Correlation\t: '+str(cv2.compareHist(hist, hist2, cv2.HISTCMP_CORREL)))
print('Chi-Square\t: '+str(cv2.compareHist(hist, hist2, cv2.HISTCMP_CHISQR)))
print('Intersection\t: '+str(cv2.compareHist(hist, hist2, cv2.HISTCMP_INTERSECT)/10))
print('Bhattacharyya\t: '+str(cv2.compareHist(hist, hist2, cv2.HISTCMP_BHATTACHARYYA)))
print('')
print('Citra 1 dengan Citra 3')
print('Correlation\t: '+str(cv2.compareHist(hist, hist3, cv2.HISTCMP_CORREL)))
print('Chi-Square\t: '+str(cv2.compareHist(hist, hist3, cv2.HISTCMP_CHISQR)))
print('Intersection\t: '+str(cv2.compareHist(hist, hist3, cv2.HISTCMP_INTERSECT)/10))
print('Bhattacharyya\t: '+str(cv2.compareHist(hist, hist3, cv2.HISTCMP_BHATTACHARYYA)))
print('')
print('Citra 2 dengan Citra 3')
print('Correlation\t: '+str(cv2.compareHist(hist2, hist3, cv2.HISTCMP_CORREL)))
print('Chi-Square\t: '+str(cv2.compareHist(hist2, hist3, cv2.HISTCMP_CHISQR)))
print('Intersection\t: '+str(cv2.compareHist(hist2, hist3, cv2.HISTCMP_INTERSECT)/10))
print('Bhattacharyya\t: '+str(cv2.compareHist(hist2, hist3, cv2.HISTCMP_BHATTACHARYYA)))
```
